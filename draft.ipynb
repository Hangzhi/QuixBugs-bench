```python
import os
from pathlib import Path

# Get all Python files from the python_programs directory
python_programs_dir = Path("/home/ubuntu/QuixBugs-bench/python_programs")

# Extract program names (excluding test files and node.py)
program_names = []
for file in python_programs_dir.glob("*.py"):
    filename = file.name
    # Exclude test files and node.py
    if not filename.endswith("_test.py") and filename != "node.py":
        # Remove .py extension to get program name
        program_name = filename[:-3]
        program_names.append(program_name)

# Sort the program names
program_names.sort()

# Display results
print(f"Total programs: {len(program_names)}\n")
print("Program names:")
print("-" * 30)
for name in program_names:
    print(name)

# If you want as a list
print("\nAs a Python list:")
print(program_names)

# If you want as a dictionary with placeholder values
programs_dict = {name: None for name in program_names}
print("\nAs a dictionary (with None values):")
print(programs_dict)
```
